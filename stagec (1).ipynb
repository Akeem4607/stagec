```python
import pandas as pd
url=r"C:\Users\akiny\Downloads\WA_Fn-UseC_-Telco-Customer-Churn.csv"
dt=pd.read_csv(url)
from matplotlib import pyplot as pt
from sklearn.preprocessing import StandardScaler,OneHotEncoder
ohe= OneHotEncoder(sparse =False)
from sklearn.model_selection import train_test_split
tts=train_test_split
import seaborn as sb
import numpy as np
pd.set_option('display.max_columns',None)
print(dt.shape)
dt.head()
```

```python
#dt['TotalCharges']=dt['TotalCharges'].astype(float)
dt['TotalCharges'] = pd.to_numeric(dt['TotalCharges'], errors='coerce')
binary_map = {'Yes': 1, 'No': 0}
dt['Churn'] = dt['Churn'].map(binary_map)
dt.head()
```

```python
y=dt['Churn']
X=dt.drop(columns='Churn')
X['TotalCharges']=X['TotalCharges'].fillna(0)
xtr,xt,ytr,yt=tts(X,y,test_size=0.2,random_state=1)
```

```python
cat_col=['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService','OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies','Contract', 'PaperlessBilling', 'PaymentMethod']
cat_tr=xtr[cat_col]
cat_t=xt[cat_col]
num_col= ['tenure', 'MonthlyCharges', 'TotalCharges']
num_tr=xtr[num_col]
num_t=xt[num_col]
```

```python
oh_tr=ohe.fit_transform(cat_tr)
oh_t=ohe.transform(cat_t)

colum = ohe.get_feature_names_out()
c_tr=pd.DataFrame(oh_tr,columns=colum)
c_t=pd.DataFrame(oh_t,columns=colum)
```

```python

stc=StandardScaler()
m_tr=stc.fit_transform(num_tr)
m_t=stc.transform(num_t)

n_tr=pd.DataFrame(m_tr,columns=num_col)
n_t=pd.DataFrame(m_tr,columns=num_col)

Xtr=pd.concat([c_tr,n_tr],axis=1,join='inner')
Xt=pd.concat([c_t,n_t],axis=1,join='inner')

```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
rfc= RandomForestClassifier(random_state=1)
etc=ExtraTreesClassifier(random_state=1)
xgb= XGBClassifier(random_state=1)
lgb=LGBMClassifier(random_state=1)
```

```python
rfc.fit(Xtr,ytr)
etc.fit(Xtr,ytr)
xgb.fit(Xtr,ytr)
lgb.fit(Xtr,ytr)
```

```python
rfc.score(Xt,yt)
```

```python
xgb.score(Xt,yt)
```
